```python
Country_names=final_df.Country.value_counts().index
Country_values=final_df.Country.value_counts().values
```

```python
## plot pie chart
plt.pie(Country_values,labels=Country_names)
```

```python
## plot pie chart for top 5 countries
plt.pie(Country_values[:5],labels=Country_names[:5])
```

```python
## plot pie chart for top 3 countries (this time displaying %)
plt.pie(Country_values[:3],labels=Country_names[:3],autopct='%1.2f%%')
### here autopct='%1.2f%%' function is used for displaying percentage on the pie chart
```

Observation: Maximum records and transactios are from India after that USA and then United Kingdom.

Observations
1. When Rating is between 4.5 to 4.9---> Excellent
2. When Rating are between 4.0 to 3.4--->very good
3. when Rating is between 3.5 to 3.9----> good
4. when Rating is between 3.0 to 3.4----> average
5. when Rating is between 2.5 to 2.9----> average
6. when Rating is between 2.0 to 2.4----> Poor
